```python
import numpy as np
import pandas as pd
#function to find the outliers in the column
def detect_outlier(data): 
    threshold=3
    outliers=[]
    mean1 = np.mean(data)
    std1 =np.std(data)    
    for y in data:
        z_score= (y - mean1)/std1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
        
#function to find the contribution of outliers to the mean
def mean_median(data):
    x=detect_outlier(data)
    y=list(filter(lambda x: not(np.isnan(x)), data))
    wo_outliers_mean=abs(abs(np.sum(y))-abs(np.sum(x)))/(len(y)-len(x))
    return abs(abs(np.mean(y))-abs(wo_outliers_mean))/abs(np.mean(y))>0.3
```

```python
#defining knn and weighted knn models
from collections import Counter
from math import sqrt

def euclidean_distance(x1, x2):
        l=[]
        for i in range(len(x1)):
            l.append((x1.iloc[i]-x2.iloc[i])**2)
        return sqrt(sum(l))
def KNN(X_train,y_train,x_test,k=3):
    y_pred=[]
    for j in range(len(x_test)):
            distances = [euclidean_distance(x_test.iloc[j], X_train.iloc[i]) for i in range(len(X_train))]
            # Sort by distance and return indices of the first k neighbors
            k_idx = np.argsort(distances)[:k]
            # Extract the labels of the k nearest neighbor training samples
            k_neighbor_labels = [y_train.iloc[i] for i in k_idx]  
            # return the most common class label
            most_common = Counter(k_neighbor_labels).most_common(1)
            y_pred.append(most_common[0][0])
    y_pred = np.array(y_pred)
    return y_pred
def weighted_KNN(X_train,y_train,x_test,k=3):
    y_pred=[]
    for j in range(len(x_test)):
            distances = [euclidean_distance(x_test.iloc[j], X_train.iloc[i]) for i in range(len(X_train))]
            # Sort by distance and return indices of the first k neighbors
            k_idx = np.argsort(distances)[:k]
            # Extract the labels of the k nearest neighbor training samples
            k_neighbor_labels = [y_train.iloc[i] for i in k_idx]  
            # return the most common class label
            d=dict()
            for i in range(len(k_neighbor_labels)):
                if k_neighbor_labels[i] not in d.keys():
                    d[k_neighbor_labels[i]]=1/k_idx[i]
                else:
                    d[k_neighbor_labels[i]]+=1/k_idx[i]
            most=max(d.items(),key=lambda x:x[1])
            y_pred.append(most[0])
    y_pred = np.array(y_pred)
    return y_pred
```

```python
#Naive Bayes Implementation

from math import sqrt
from math import pi
from math import exp

# Calculate the mean, median, and number of rows as tuples for each column belonging to a specific class 
def summarize_by_class(df1):
    summaries = dict()
    for i in df1['P. Habitable Class'].unique():
        is_i=df1['P. Habitable Class']==i
        df_temp=df1[is_i]
        summaries[i]=list()
        for j in df_temp:
            mean=np.mean(df_temp[j])
            std=np.std(df_temp[j])
            n=len(df_temp)
            summaries[i].append((mean,std,n))
    return summaries


# Calculate the Gaussian probability distribution function for x
def calculate_probability(x, mean, stdev):
        if(stdev):
            exponent = exp(-((x-mean)**2 / (2 * stdev**2 )))
            return (1 / (sqrt(2 * pi) * stdev)) * exponent
        else:
            return 0

# Calculate the probabilities of predicting each class for a given row
def calculate_class_probabilities(summaries, row):
    total_rows = sum([summaries[label][0][2] for label in summaries])
    probabilities = dict()
    for class_value, class_summaries in summaries.items():
        #print(len(class_summaries))
        probabilities[class_value] = summaries[class_value][0][2]/float(total_rows)
        for i in range(len(class_summaries)):
            mean, stdev, _ = class_summaries[i]
            probabilities[class_value] *= calculate_probability(row[i], mean, stdev)
    return probabilities

# Predict the class for a given row
def predict(summaries, row):
    probabilities = calculate_class_probabilities(summaries, row)
    best_label, best_prob = -1, -1
    for class_value, probability in probabilities.items():
        if(best_label==(-1) or probability > best_prob):
            best_prob = probability
            best_label = class_value
    return best_label

# Naive Bayes Algorithm
def naive_bayes(train, test):
    summarize = summarize_by_class(train)
    predictions = list()
    for row in range(len(test)):
        output = predict(summarize, test.iloc[row])
        predictions.append(output)
    return(predictions)
```

```python
#Column Names
X_col=['P. Mass (EU)', 'P. Radius (EU)', 'P. Density (EU)', 'P. Gravity (EU)','P. Eccentricity', 'P. Mean Distance (AU)','P. Esc Vel (EU)','P. HZD', 'P. HZC', 'P. HZA', 'P. HZI', 'P. ESI', 'P_TsMax', 'P_TsMin', 'P_TsMean']
```

```python
#DECISION TREES

def entropy_func(cnt, num):
    return -(cnt*1.0/num)*math.log(cnt*1.0/num, 2)

#return entropy of a group 
#cnt1 and cnt2 are counts of two classes
def entropy_cal(cnt1, cnt2):
    #only one class in the group, entropy is 0
    if cnt1== 0 or cnt2 == 0:  
        return 0
    return entropy_func(cnt1, cnt1+cnt2) + entropy_func(cnt2, cnt1+cnt2)

#get the entropy of the divided group
def entropy_of_one_division(division): 
   
    entropy = 0
    n = len(division)
    classes = set(division)
    for c in classes:   # for each class, get entropy
        n_c = sum(division==c)
        e = n_c*1.0/n * entropy_cal(sum(division==c), sum(division!=c)) # weighted avg
        entropy += e
    return entropy, n

# The whole entropy of two big circles combined
#get the entropy of split
#y_predict is the split decision
def get_entropy(y_predict, y_real):
    if len(y_predict) != len(y_real):
        print('variables are not of same length!')
        return None
    n = len(y_real)
    entropy_true, n_true = entropy_of_one_division(y_real[y_predict]) # left hand side entropy
    entropy_false, n_false = entropy_of_one_division(y_real[~y_predict]) # right hand side entropy
    entropy = entropy_true*1.0/n * entropy_true + entropy_false*1.0/n * entropy_false # overall entropy, again weighted average
    return entropy

class DecisionTreeClassifier(object):
    def __init__(self, max_depth):
        self.depth = 0
        self.max_depth = max_depth
    
    def fit(self, x, y, root_node={}, depth=0):
        if root_node is None: 
            return None
        elif len(y) == 0:
            return None
        elif self.all_same(y):
            return {'val':y[0]}
        elif depth >= self.max_depth:
            return None
        else: 
            col, cutoff, entropy = self.find_best_split(x, y)    # find one split given an information gain 
            y_left = y[x[:, col] < cutoff]
            y_right = y[x[:, col] >= cutoff]
            root_node = {'col': X_col[col], 'col_index':col,
                        'cutoff':cutoff,
                       'val': np.round(np.mean(y))}
            root_node['left'] = self.fit(x[x[:, col] < cutoff], y_left, {}, depth+1)
            root_node['right'] = self.fit(x[x[:, col] >= cutoff], y_right, {}, depth+1)
            self.depth += 1 
            self.trees = root_node
            return root_node
    
    def find_best_split(self, x, y):
        col = None
        min_entropy = 1
        cutoff = None
        for i, c in enumerate(x.T):
            entropy, cur_cutoff = self.split(c, y)
            if entropy == 0:    # find the first perfect cutoff. Stop Iterating
                return i, cur_cutoff, entropy
            elif entropy <= min_entropy:
                min_entropy = entropy
                col = i
                cutoff = cur_cutoff
        return col, cutoff, min_entropy
    
    def split(self, col, y):
        min_entropy = 10
        n = len(y)
        for value in set(col):
            y_predict = col < value
            my_entropy = get_entropy(y_predict, y)
            if my_entropy <= min_entropy:
                min_entropy = my_entropy
                cutoff = value
        return min_entropy, cutoff
    
    def all_same(self, items):
        return all(x == items[0] for x in items)
                                           
    def predict(self, x):
        tree = self.trees
        results = np.array([0]*len(x))
        for i, c in enumerate(x):
            results[i] = self.predict_row(c)
        return results
    
    def predict_row(self, row):
        cur_node = self.trees
        
        while (cur_node and cur_node.get('cutoff')):
            if(cur_node['left']==None and cur_node['right']==None):
                return cur_node.get('val')
            if row[cur_node['col_index']] < cur_node['cutoff']:
                if(cur_node['left']==None):
                    return cur_node.get('val')
                else:
                    cur_node = cur_node['left']
            else:
                if(cur_node['right']==None):
                    return cur_node.get('val')
                else:
                    cur_node = cur_node['right']
        return cur_node.get('val')
```

```python
from collections import Counter 
def mode_custom(n_num):
    n = len(n_num) 
    data = Counter(n_num) 
    get_mode = dict(data) 
    mode = [k for k, v in get_mode.items() if v == max(list(data.values()))] 
    if(len(mode) == n): 
        get_mode = mode[0]
    else: 
        get_mode = max(mode) 
      
    return get_mode
```

```python
# RANDOM FOREST
import statistics
class RandomForest():
    def __init__(self, x, y, n_trees, sample_sz, depth):
        np.random.seed(12)
          
        self.x, self.y, self.sample_sz, self.depth  = x, y, sample_sz, depth
        
        self.objects=[]
        self.trees=[]
        for i in range(n_trees):
            o,t = self.create_tree()
            self.objects.append(o)
            self.trees.append(t)
        
        
    def create_tree(self):
        idxs = np.random.permutation(len(self.y))[:self.sample_sz]
        #f_idxs = np.random.permutation(self.x.shape[1])[:self.n_features]
        clf=DecisionTreeClassifier(self.depth)
        return clf,clf.fit(self.x[idxs],self.y[idxs])
        
    def predict(self, x):
        tot_pred=pd.DataFrame()
        i=1
        for t in self.objects:
            tot_pred[i]=t.predict(x)
            i+=1        
        return tot_pred.mode(axis=1)[0]

def std_agg(cnt, s1, s2): return math.sqrt((s2/cnt) - (s1/cnt)**2)
```
